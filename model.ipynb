```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Data_for_training.csv')
```

```python
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
df.plot(kind='scatter', grid=True, x='GPD', y='BLI')
plt.axis([20000, 80000, 4, 9])
```

```python
model = LinearRegression()
```

```python
X=df[['GPD']]
y = df[['BLI']]
```

```python
model.fit(X, y)
```

```python
X_new = [[37655]]
model.predict(X_new)
```

```python
plt.scatter(df['GPD'], df['BLI'], color='blue', label='Données originales')

y_pred = model.predict(X)
plt.plot(df['GPD'], y_pred, color='red', label='Ligne de régression')
plt.grid(True)
plt.title('Relation entre GPD et BLI avec Ligne de Régression')
plt.xlabel('GPD (USD par habitant)')
plt.ylabel('Indice de Bonheur de Vie (BLI)')
plt.legend()
```
